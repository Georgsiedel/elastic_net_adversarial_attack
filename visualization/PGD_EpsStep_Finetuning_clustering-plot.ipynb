```python
import matplotlib.pyplot as plt
import json
```

```python
#This code is only for reproducing paper results from existing json file currently
json_file_1 = f'../data/paper_data/pgd_standard_eps_iter_3e-2.json'
json_file_2 = f'../data/paper_data/pgd_standard_eps_iter_3e-3.json'
json_file_3 = f'../data/paper_data/pgd_standard_eps_iter_3e-4.json'

json_files = [json_file_1, json_file_2, json_file_3]
adversarial_distances_dict = {}

for i, json_file in enumerate(json_files):
    with open(json_file, 'r') as f:
        data = json.load(f)

    indices = list(range(len(data['projected_gradient_descent']['runtime'])))
    adversarial_distances_dict[i] = data['projected_gradient_descent']['adversarial_distance']
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].scatter(indices, adversarial_distances_dict[0], s=15, c='blue')
ax[0].set_title('eps_step = 0.03')
ax[0].set_xlabel('ImageID')
ax[0].set_ylabel('L$_2$ Distance')
ax[0].tick_params(axis='both')
ax[1].scatter(indices, adversarial_distances_dict[1], s=15, c='blue')
ax[1].set_title('eps_step = 0.003')
ax[1].set_xlabel('ImageID')
ax[1].tick_params(axis='both')
ax[2].scatter(indices, adversarial_distances_dict[2], s=15, c='blue')
ax[2].set_title('eps_step = 0.0003', fontsize=8)
ax[2].set_xlabel('ImageID')
ax[2].tick_params(axis='both')
plt.suptitle('Fine-tuning eps_step parameter to avoid clustering in adversarial distance calculations', weight='bold', y=1.05)
plt.tight_layout()
plt.show()
```
